<h1><b>Instrumentos Piezoresistivos</b></h1>

<h2><b>Capítulo 1: Introdução</b></h2>

<br>
Em 1856, William Thomsom, também conhecido como Lorde Kelvin, publicou seus estudos a cerca da resistividade elétrica para materiais metálicos. Em suas observações, concluiu que a <b>resistência elétrica</b> de um metal varia diretamente com o comprimento do condutor e inversamente com relação à área de sua seção reta, bem como variava diretamente com relação a uma constante que chamou de <b>resistividade elétrica</b>.</br>
$$
\\
$$
\begin{equation*}
R = \frac{r \cdot L}{A}
\end{equation*}

Onde:

* R é a resistência elétrica
* r é a resistividade elétrica do material
* L é o comprimento do condutor
* A é a área da seção reta do condutor

<br>A resistividade elétrica, também chamada de resistência elétrica específica, indica o quanto um material resiste a passagem de elétrons, sendo, portanto, o inverso da <b>condutividade elétrica</b>, indicada por <i>s</i>.
$$
\\
$$
\begin{equation*}
s = \frac{1}{r}
\end{equation*}
</br>

<h2><b>Capítulo 2: Variações na Resistividade Elétrica</b></h2>
<p style='text-align: justify;'>
<br>
Embora a resistividade elétrica de um material seja assumida constante em muitos casos, isso nem sempre pode ser assumido. Em geral, tais variações são suficientemente pequenas e podem ser desprezadas mas, se a aplicação for para utilização em sensores, ela torna-se inadequada. Em metais, a resistividade elétrica variará somente em função de sua <b>geometria</b>, sendo esta, por sua vez, uma função da <b>temperatura</b> e dos <b>esforços mecânicos</b>.
</br>
</p>
<h3><b>Variações na resistividade elétrica devido à temperatura</b></h3>

<br>
<p style='text-align: justify;'>
A resistividade elétrica de um material pode variar conforme aumenta-se a sua temperatura, em razão da variação da sua geometria devido à <b>dilatação volumétrica</b>. É possível, no entanto, condensar as variações triaxiais em um único parâmetro, denominado <b>coeficiente de temperatura da resistividade</b>. Assim, a variação da resistividade em função da temperatura pode ser expressa como:
$$
\\
$$
</p>    
</br>

\begin{equation*}
\Delta r = \alpha_{0}\cdot r_{0} \cdot \Delta T
\end{equation*}
<br>Onde:</br>
* $r_{0}$ é a resistividade na temperatura de referência [$\frac{\Omega}{m}$]
* $\alpha_{0}$ é o coeficiente de temperatura da resistividade [$\frac{1}{K}$]

<br> Reescrevendo a equação anterior de uma forma mais conveniente para aplicações, temos que:
$$
\\
$$
</br>
\begin{equation*}
r = r_{0} \cdot [1 - \alpha_{0} \cdot (T - T_{0})] 
\end{equation*}
<br>Onde:</br>
* T é a temperatura analisada [K]
* $T_{0}$ é a temperatura de referência [K]
* r é a resistividade elétrica na temperatura analisada [$\frac{\Omega}{m}$]

<br>
<p style='text-align: justify;'>
Usualmente, a temperatura de referência utilizada é a 20ºC, obtida experimentalmente para diversos materiais. Elas, no entanto, são apresentadas, muitas vezes, em [$\frac{1}{ºC}$], uma vez que, para os cálculos, a escala da temperatura é irrelevante, desde todas as variáveis estejam relacionadas a uma mesma escala. É possível, também, obter as resistividades elétricas de alguns materiais através de tabelas.
</br>
</p>

<h3><b>Variações na resistividade elétrica devido a esforços mecânicos</b></h3>
<p style='text-align: justify;'>
<br>
Materiais submetidos a esforços mecânicos estão constantemente tendo sua geometria alterada devido às <b>deformações</b>, que podem alterar a geometria da peça de várias formas diferentes. Tais deformações podem ser captadas por um sensor chamado <b>extensômetro</b>, cujas propriedades elétricas são alteradas devido às <b>solicitações mecânicas</b>, havendo, portanto, uma relação <b>eletro-mecânica</b> entre as propriedades do material e que permitem a medição das deformações da peça. Uma vez que se conhece a deformação, é possível, via <b>lei de Hooke</b> determinar as <b>tensões mecânicas</b> atuantes. Devido a relevância dessas aplicações, esses extensômetros são também conhecidos como <b>strain gauges</b>: <i>sensores de deformação</i>.
</br>
<br>

De forma matemática, a relação entre variação da resistência elétrica e deformação pode ser representada como:
$$
\\
$$
\begin{equation*}
GF = \frac{\frac{\Delta R}{R_{0}}}{\epsilon}   \;\;\;\;\;\; (1)
\end{equation*}
Sendo:
* GF o fator gauge, ou fator do extensômetro
* $\epsilon$ a deformação


<h2><b>Capítulo 3: Ponte de Wheatstone</b></h2>
<p style='text-align: justify;'>
<br>
Como visto acima, os extensômetros fornecem dados de resistência elétrica e, a partir da sua variação, é calculada a deformação do material. Medir diretamente a resistência elétrica, no entanto, é pouco prático; de forma geral, as <b>aquisições de dados</b> são feitas medindo tensão e corrente elétricas. Assim, para realizar essa transformação, utiliza-se uma <b>ponte de Wheatstone</b>. Nela, dois pares de resistores são montados em paralelo, sendo cada resistor de cada par montado em série, conforme ilustra o diagrama abaixo.
</br>
</p>
$$
\\
$$
<center><b>Figura 3.</b> Diagrama elétrico de uma ponte de Wheatstone.</center>
$$
\\
$$
<center><img src="Figuras/Acelerômetros Piezoresistivos/Figura 3.png" style="width: 250px;"></center>
$$
\\
$$
<p style='text-align: justify;'>
<br>
Na <b>figura 3</b>, os resistores $R_{1}$, $R_{2}$ e $R_{3}$ são conhecidos, sendo $R_{x}$ o resistor (nesse caso o extensômetro) cuja resistência é desconhecida. Uma fonte de tensão elétrica tem seus terminais acoplados aos nós <i>A</i> e <i>D</i> e a diferença de tensão elétrica entre os nós <i>B</i> e <i>C</i> é medida. Via <b>lei de Ohm</b> é possível determinar o valor da resistência elétrica do extensômetro, como demonstrado abaixo.
</br>
</p>
Conhecendo a tensão fornecida pela fonte de tensão elétrica (<i>$V_{fonte}$</i>), podemos determinar a tensão elétrica do nó <i>C</i> (<i>$V_{C}$</i>) como:
$$
\\
$$
\begin{equation*}
V_{C} = \frac{R_{2}}{R_{2} + R_{1}} \cdot V_{fonte}   \;\;\;\;\;\; (2)
\end{equation*}
$$
\\
$$
Da mesma forma, pode-se calcular o valor da tensão elétrica no nó <i>B</i> (<i>$V_{B}$</i>) como:
$$
\\
$$
\begin{equation*}
V_{B} = \frac{R_{3}}{R_{3} + R_{x}} \cdot V_{fonte}   \;\;\;\;\;\; (3)
\end{equation*}
$$
\\
$$
Sendo <i>$V_{C-B}$</i> a diferença de tensão elétrica entre os nós <i>C</i> e <i>B</i>, temos que:
$$
\\
$$
\begin{equation*}
V_{C-B} = V_{C} - V_{B}   \;\;\;\;\;\; (4)
\end{equation*}
$$
\\
$$
\begin{equation*}
V_{C-B} = \frac{R_{2}}{R_{2} + R_{1}} \cdot V_{fonte} - \frac{R_{3}}{R_{3} + R_{x}} \cdot V_{fonte}
\end{equation*}
$$
\\
$$
\begin{equation*}
V_{C-B} = V_{fonte} \cdot \biggl(\frac{R_{2}}{R_{2} + R_{1}} - \frac{R_{3}}{R_{3} + R_{x}}\biggl)
\end{equation*}
$$
\\
$$
Reescrevendo a equação acima, isolando a $R_{x}$ a fim de determiná-lo mais facilmente e considerando que todas os demais valores são conhecidos, temos que:
$$
\\
$$
\begin{equation*}
R_{x} = \frac{R_{3} \cdot V_{fonte} \cdot (R_{2} + R_{1})}{R_{2} \cdot V_{fonte} - V_{C-B} \cdot (R_{2} + R_{1})} - R_{3}   \;\;\;\;\;\; (5)
\end{equation*}
$$
\\
$$
<p style='text-align: justify;'>
<br>
No diagrama apresentado na figura 3, a resistência $R_{2}$ é indicado com o símbolo de uma flecha. Tal símbolo representa um resistor de resistência variável, ou seja, um <b>potenciômetro</b>. A sua função no circuito é a de realizar o <b>balanceamento</b> do mesmo, ou seja, que a diferença de tensão elétrica entre os nós <i>C</i> e <i>B</i> seja igual a zero ($V_{C-B}$ = 0) para uma dada condição de referência que, neste caso, é quando o extensômetro não está submetido a esforços mecânicos. O balanceamento, embora seja recomendado para muitas situações, nem sempre é utilizado, ficando a critério do usuário balancear ou não o circuito. A <b>figura 4</b> ilustra a variação de tensão elétrica entre os nós <i>C</i> e <i>B</i> enquanto um extensômetro é deformado em uma ponte de Wheatstone balanceada.

<h3><b>Meia ponte de Wheatstone</b></h3>
<p style='text-align: justify;'>
<br>
Uma meia ponte de Wheatstone utiliza dois resistores ativos, ou seja, ambos são acoplados ao corpo de prova e medem as deformações simultaneamente. Se forem acoplados à mesma face do corpo de prova, um paralelo ao eixo longitudinal e outro perpendicular a ele, é capaz de realizar medidas tanto de tração e compressão quanto de flexão, sendo chamada de configuração I. Nela, um dos extensômetros compensa o efeito de Poisson, enquanto o outro realiza as medidas de interesse. Quando posiciona um extensômetro paralelos entre si, mas em faces opostas, o circuito é capaz de medir somente flexão, com cada um dos extensômetros podendo alternar entre medidas de compressão e de tração, sendo chamada de configuração II. Em ambos os casos os efeitos de variação de temperatura são compensados. A <b>figura 6</b> ilustra ambas as configurações.

<h3><b>Ponte completa de Wheatstone</b></h3>
<p style='text-align: justify;'>
<br>
A ponte completa de Wheatstone é composta por quatro extensômetros ativos e que podem ser dispostos em três configurações distintas. Todas as três configurações conseguem compensar os efeitos de deformação decorrentes da variação de temperatura. <h2><b>Capítulo 4: Extensômetros do tipo roseta</b></h2>

<p style='text-align: justify;'>
<br>
Como mencionado anteriormente, extensômetros somente conseguem medir <b>deformações em uma única direção</b>. Nas configurações de ponte de Wheatstone apresentadas no <b>capítulo 3</b>, aquelas que permitiam a medição de deformação axial e de flexão eram recomendadas apenas para carregamentos com apenas um dos esforços, não com os dois simultaneamente, uma vez que isso acarretaria em um <b>erro de medida</b>. Os usos dessas configurações, portanto, é limitado a componentes cujos carregamentos na face em que o extensômetro é posicionado se devam a um único tipo de esforço. Para carregamentos mais complexos, como o biaxial, recomenda-se o uso de <b>extensômetros do tipo roseta</b>.
</br>
</p>

<p style='text-align: justify;'>
<br>
Tais extensômetros são idênticos àquelas que medem deformações uniaxiais, o que muda é a sua configuração geométrica: se para os extensômetros uniaxiais os strain gauges eram dispostos sempre em ângulos retos ou paralelos à direção do esforço, nos extensômetros do tipo roseta os strain gauges são posicionadas em ângulo com o esforço. Isso, naturalmente, poderia acarretar em um erro de medida caso usássemos um único strain gauge; utilizando-se dois ou mais, no entanto, é possível montar um sistema de equações que permite a obtenção do <b>estado plano de tensões</b> na face do material instrumentado. Para tanto, é necessário antes compreender os tipos principais de extensômetros roseta: os empilhados e os planares. Os extensômetros roseta também podem fazer uso de pontes de Wheatstone em cada um dos seus elementos, tal como os uniaxias, fazendo-se uso especialmente da configuração de um quarto de ponte, para balancear o circuito.
</br>
</p>

<h3><b>Extensômetros Roseta: Empilhados</b></h3>
<p style='text-align: justify;'>
<br>
Estes extensômetros são utilizados principalmente quando se conhecem as direções das tensões principais de um carregamento, de modo que seja necessário o uso de apenas dois straing gauges para determinação dos esforços principais. Os dois são posicionados perpendiculares entre si e concidentes às direções dos esforços, como ilustra a <b>figura 8</b>.
</br>

<center><b>Figura 8.</b> Posicionamento de dois extensômetros do tipo roseta empilhados para medição das tensões principais.</center>
$$
\\
$$
<center><img src="Figuras/Acelerômetros Piezoresistivos/Figura 8.png",style="width: 400px;"></center>

<p style='text-align: justify;'>
<br>
Assim, sendo $\nu$ o coeficiente de Poisson, E o módulo de elasticidade, com $\varepsilon_{1}$ e $\varepsilon_{2}$ como sendo as deformações nos eixos 1 e 2, as tensões principais $\sigma_{1}$ e $\sigma_{2}$ podem ser calculadas utilizando-se as relações abaixo: 
</br>
</p>   
$$
\\
$$
\begin{equation*}
\sigma_{1} = \frac{E}{1-\nu^{2}}(\varepsilon_{1} + \nu \varepsilon_{2}) \;\;\;\;\;\; (6)
\end{equation*}
$$
\\
$$
\begin{equation*}
\sigma_{2} = \frac{E}{1-\nu^{2}}(\varepsilon_{2} + \nu \varepsilon_{1}) \;\;\;\;\;\; (7)
\end{equation*}
</p>
<p style='text-align: justify;'>
<br>
Esta configuração também pode ser utilizada para determinação de esforços cujas direções principais sejam desconhecidas, mas não é aconselhável, uma vez que apresentam dificuldade de dissipar calor e, portanto, sofrem maior influência da temperatura. Ainda, os extensômetros roseta empilhados possuem baixa sensibilidade às deformações, uma vez que somente o primeiro elemento está em contato com a estrutura, estando os demais strain gauges à mercê da transmissão das deformações do primeiro. Somente são aconselháveis caso o espaço disponível seja limitado ou caso deseje-se a medição de tensões pontuais muito elevadas; para estes casos, utiliza o equacionamento apresentado no tópico abaixo, sobre <b>extensômetros roseta planares</b>; o seu posicionamento, portanto, é tal o exemplificado no tópico de rosetas planares.
</br>
</p>

<h3><b>Extensômetros Roseta: Planares</b></h3>
<p style='text-align: justify;'>
<br>
Esta configuração é o tipo mais comumente utilizado para determinação de esforços biaxiais. Nela, os extensômetros são posicionados espaçada entre si, com ângulos constantes, como mostra a <b>figura 9.</b>
</br>
</p>
<center><b>Figura 9.</b> Posicionamento de três extensômetros do tipo roseta planar para medição das tensões principais.</center>
$$
\\
$$
<center><img src="Figuras/Acelerômetros Piezoresistivos/Figura 9.png",style="width: 400px;"></center>
$$
\\
$$

Cada um dos strain gauges fornecerá um resultado para deformação, cujas relações com os eixos x e y são mostradas abaixo, de forma que seja possível elaborar um sistema de equações para determinação das deformações $\varepsilon_{x}$ e $\varepsilon_{y}$.
$$
\\
$$
\begin{equation*}
\varepsilon_{A} = \varepsilon_{x}cos^{2}(\alpha_{A}) + \varepsilon_{x}sen^{2}(\alpha_{A}) + \gamma_{xy}cos(\alpha_{A})sen(\alpha_{A}) \;\;\;\;\;\; (8)
\end{equation*}
$$
\\
$$
\begin{equation*}
\varepsilon_{B} = \varepsilon_{x}cos^{2}(\alpha_{B}) + \varepsilon_{x}sen^{2}(\alpha_{B}) + \gamma_{xy}cos(\alpha_{B})sen(\alpha_{B}) \;\;\;\;\;\; (9)
\end{equation*}
$$
\\
$$
\begin{equation*}
\varepsilon_{C} = \varepsilon_{x}cos^{2}(\alpha_{C}) + \varepsilon_{x}sen^{2}(\alpha_{C}) + \gamma_{xy}cos(\alpha_{C})sen(\alpha_{C}) \;\;\;\;\;\; (10)
\end{equation*}
$$
\\
$$

Sendo:

* $\varepsilon_{A}$ a deformação no strain gauge A
* $\varepsilon_{B}$ a deformação no straing gauge B
* $\varepsilon_{C}$ a deformação no straing gauge C
* $\varepsilon_{x}$ a deformação na direção x
* $\varepsilon_{y}$ a deformação na direção y
* $\alpha_{A}$ o ângulo do strain gauge A
* $\alpha_{B}$ o ângulo do strain gauge B
* $\alpha_{C}$ o ângulo do strain gauge C

$$
\\
$$

Uma vez calculadas as deformações nos eixos x e y, $\varepsilon_{x}$ e $\varepsilon_{y}$ respectivamente, pode-se calcular as deformações principais, $\varepsilon_{1}$ e $\varepsilon_{2}$, conforme mostrado abaixo:

\begin{equation*}
\varepsilon_{1} = \frac{1}{2}(\varepsilon_{x} + \varepsilon_{y}) + \frac{1}{2}\sqrt{(\varepsilon_{x} - \varepsilon_{y})^{2} + \gamma_{xy}^{2}}  \;\;\;\;\;\; (11)
\end{equation*}

$$
\\
$$

\begin{equation*}
\varepsilon_{2} = \frac{1}{2}(\varepsilon_{x} + \varepsilon_{y}) - \frac{1}{2}\sqrt{(\varepsilon_{x} - \varepsilon_{y})^{2} + \gamma_{xy}^{2}}  \;\;\;\;\;\; (12)
\end{equation*}

$$
\\
$$

\begin{equation*}
tan(2 \theta) = \frac{\gamma_{xy}}{\varepsilon_{x} - \varepsilon_{y}} \;\;\;\;\;\; (13)
\end{equation*}
$$
\\
$$
<p style='text-align: justify;'>
<br>
Onde $\theta$ pode assumir os valores de $\theta_{1}$, que é o ângulo entre a máxima deformação principal, $\varepsilon_{1}$, e o eixo x, ou $\theta_{2}$, que é o ângulo entre a mínima deformação principal, $\varepsilon_{2}$, com relação ao eixo x. Resolvendo as equações (11), (12) e (13), obtém-se os resultados das deformações principais, cujos valores podem ser substituídos nas equações (6) e (7) a fim de encontrar as tensões principais.

<h3><b>Exemplo</b></h3>
<p style='text-align: justify;'>
<br>
Em uma fábrica de peças de automóveis, o capô de um dos modelos é fabricado através da estampagem de uma chapa de alumínio Al 6061 utilizando-se uma prensa hidráulica. Para o controle desta prensa, uma célula de carga piezorresistiva utilizando extensômetros roseta planares é instalada, de forma que seja possível conhecer as tensões biaxiais exercidas sobre a chapa e que, deste modo, possa se determinar a máxima tensão a ser aplicada pela prensa. Sabendo que cada um dos três extensômetros, A, B e C, utiliza a configuração de um quarto de ponte de Wheatstone, em que cada resistor possui resistências elétricas de $R_{1}$ = 12 k$\Omega$, $R_{2}$ = 50 $\Omega$ e $R_{3}$ = 600 $\Omega$, com tensões elétricas de $V_{C-B_{A}}$ = 1.21 $\mu$V, $V_{C-B_{B}}$ = 2.32 $\mu$V e $V_{C-B_{C}}$ = 3,15 $\mu$V, alimentados por uma fonte com tensão elétrica de $V_{fonte}$ = 5 V, e que todos os extensômetros são iguais e com fator gauge GF = 2, determine a máxima e mínima tensões mecânicas exercidas sobre a chapa neste instante. Considere que o primeiro extensômetro está posicionado a 60º com relação ao eixo x.
</br>
</p>

<p style='text-align: justify;'>
<br>
Utilizando a equação (4), considerando que a ponte está balanceada, $V_{CB}$ = 0. Dessa forma, substituindo as equações (2) e (3), obtém-se que:
</br>
</p> 

```python
R1 = 12000
R2 = 50
R3 = 600

Vfonte = 5

Vc = (R2 * Vfonte)/(R1 + R2)

R0 = (R3 * Vfonte)/(Vc) - R3

print('O valor da resistência inicial dos strain gauges é de ', R0, ' \u03A9.')
```

<p style='text-align: justify;'>
<br>
Aplicando a equação (5) para cada um dos strain gauges, é possível encontrar as resistências de cada um dos extensômetros.
</br>
</p> 

```python
VCBa = 1.21 * 10**-6
VCBb = 2.32 * 10**-6
VCBc = 3.15 * 10**-6

RxA = R3 * Vfonte * (R2 + R1)/(R2 * Vfonte - VCBa * (R2 + R1))
RxB = R3 * Vfonte * (R2 + R1)/(R2 * Vfonte - VCBb * (R2 + R1))
RxC = R3 * Vfonte * (R2 + R1)/(R2 * Vfonte - VCBc * (R2 + R1))
```

Utilizando a equação (1) para cada strain gauge, encontram-se as deformações em cada extensômetro.

```python
GF = 2

eA = (RxA - R0)/(R0 * GF)
eB = (RxB - R0)/(R0 * GF)
eC = (RxC - R0)/(R0 * GF)
```

Resolvendo o sistema das equações (8), (9) e (10), é possível encontrar as deformações $\varepsilon_{x}$, $\varepsilon_{y}$ e $\varepsilon_{xy}$ para o conjunto de três extensômetros que compõe o extensômetro roseta planar.

```python
import numpy as np

def sen(alpha):
    
    alpha = np.deg2rad(alpha)
    
    seno = np.sin(alpha)
    
    return seno

def cos(alpha):
    
    alpha = np.deg2rad(alpha)
    
    cosseno = np.cos(alpha)
    
    return cosseno

alphaA = 60
alphaB = 120
alphaC = 180

A = np.array([[cos(alphaA)**2, sen(alphaA)**2, cos(alphaA)*sen(alphaA)],
              [cos(alphaB)**2, sen(alphaB)**2, cos(alphaB)*sen(alphaB)],
              [cos(alphaC)**2, sen(alphaC)**2, cos(alphaC)*sen(alphaC)]])

b = np.array([eA, eB, eC])

x = np.linalg.solve(A,b)

ex, ey, exy = x[0], x[1], x[2]

print(ex, ey, exy)
```

Substituindo os valores encontrados para $\varepsilon_{x}$, $\varepsilon_{y}$ e $\varepsilon_{xy}$, e substituindo-os nas equações (11) e (12), encontram-se as deformações máxima e mínima, $\varepsilon_{1}$, $\varepsilon_{2}$, respectivamente. Substituindo esses valores nas equações (6) e (7), encontram-se os valores para $\sigma_{1}$ e $\sigma_{2}$.

```python
e1 = 0.5*(ex + ey) + 0.5*np.sqrt((ex - ey)**2 + exy**2)
e2 = 0.5*(ex + ey) - 0.5*np.sqrt((ex - ey)**2 + exy**2)

#angle = np.arctan(np.deg2rad(exy/(ex - ey)))/2

v = 0.33
E = 68.9*10**9

sigma1 = E*(e1 + v*e2)/(1 - v**2)

sigma2 = E*(e2 + v*e1)/(1 - v**2)

print('A tensão máxima é de ', sigma1/10**6, ' MPa. A tensão mínima é de ', sigma2/10**6, ' Mpa.')
```
