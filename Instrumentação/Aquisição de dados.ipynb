<h1><b>Aquisição de Dados</b></h1>

<h2><b>Introdução</b></h2>
<p style='text-align: justify;'>
<br>A <b>aquisição de dados</b> pode ser traduzida, de forma sintética, como a transformação de informações analógicas em informações que possam ser compreendidas pelo computador, i.e., transformar um <b>sinal analógico</b> em um <b>sinal digital</b>, de forma que possa ser armazenado para posterior análise mediante o uso de métodos computacionais.</br>
<br>A aquisição de dados, também chamada de DAQ, é um sistema formado pela junção de transdutores, um <i>hardware</i> (ou placa DAQ) e um <i>software</i> para monitorar e manipular os dados adquiridos de forma adequada.
</p>
    
<h2><b>Capítulo 1: Sinais</b></h2>
<p style='text-align: justify;'>
<br>O termo "sinais" pode conter vários significados a depender da área de estudo, podendo representar ações, funções e afins. De modo geral, um sinal carrega consigo uma <b>informação</b> que possa ser interpretada por algo ou alguém. Nesse contexto, medir o nível de um observatório com uma régua e registrar os dados obtidos pode ser interpretado como um sinal. Nessa apostila, no entanto, procura-se um sentido menos amplo. Dessa forma, um sinal será entendido como uma entidade, usualmente uma onda eletromagnética, que transporte informação sobre um <b>fenômeno físico</b> e que seja <b>função</b> de uma ou mais variáveis, podendo estar contidas no domínio do tempo, i.e. contínuo, ou no dominínio da frequência, i.e. discreto.  
</p>
<h3><b>Sinais analógicos</b></h3>
<p style='text-align: justify;'>
<br>Sinais analógicos são sinais adquiridos por um <b>transdutor</b>. Transdutores são instrumentos que medem uma determinada grandeza utilizando um <b>sensor</b> e que fornecem uma grandeza de saída diferente da de entrada, mas cujas grandezas são inter-relacionadas. De modo geral, para <b>medições mecânicas</b>, as grandezas são medidas utilizando um transdutor, como uma célula de carga, cujas medições se baseiam no uso de correlações entre as <b>propriedades eletro-mecânicas</b> do instrumento de medida, sendo usualmente essa propriedade a tensão elétrica.</br>

<h3><b>Sinais digitais</b></h3>
<br>Sinais digitais são sinais cujos valores são representados como <b>números binários</b>. Um número binário é composto pela combinação dos números 0 e 1 que representa um <b>número decimal</b>. A transformação de um valor decimal para um valor binário se dá no <i>hardware</i> de aquisição de dados (placa DAQ) por meio de um circuito elétrico denominado <b>conversor analógico-digital</b>. Nele, um sinal analógico é transformado para o domínio da frequência, sendo representados por uma sequência de <b>pulsos</b> que representam um número binário.
</p>

<h2><b>Capítulo 2: Transdutores</b></h2>
<p style='text-align: justify;'>
<br>Transdutores e sensores são instrumentos de medida que, embora sejam utilizados como sinônimos, são conceitos diferentes. Embora não haja um consenso entre autores, o VIM (Vocabulário Internacional de Metrologia) define um sensor como sendo um elemento presente num sistema de medição que é afetado diretamente por um fenômeno físico. Um transdutor, por outro lado, é um dispositivo de medidas que é sensível a um fenômeno, captado por um sensor, e cuja grandeza de saída possui uma relação especificada com a grandeza do fenômeno captado (entrada). As grandezas de saída, usualmente, são a tensão ou corrente elétricas, na forma de <b>sinais</b>.</br>
<br>Embora existam correlações eletro-mecânicas bem estabelecidas e matematicamente consistentes, os cálculos para converter os dados em função de propriedades elétricas fornecidas pelo transdutor em dados mecânicos de interesse nem sempre são triviais. Dessa forma, é usual que sejam utilizadas <b>funções de transferência</b> que transformem uma propriedade elétrica em uma propriedade mecânica de uma forma mais simples, sendo essa função definida experimentalmente e que deve ser fornecida pelo fabricante do transdutor.</br>

<h3>Exemplo:</h3>
<br>Considere o seguinte sinal analógico de tensão elétrica variante no tempo obtido por uma célula de carga acoplada a um eixo sob rotação.
</p>

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,20,0.001)

sinal_V = 0.006*np.sin(2*t) + 0.0015*np.cos(2*t) + 0.05*np.sin(6*t) + 0.02*np.sin(6*t) + 0.25*np.sin(8*t) + 0.15*np.cos(8*t)

plt.plot(t, sinal_V)
plt.xlabel('Tempo (s)')
plt.ylabel('Volts (V)')
plt.title('Sinal adquirido por uma célula de carga acoplada a um eixo sob rotação')
```

<p style='text-align: justify;'>
<br>Considere, agora, que a função de transferência dessa célula de carga é dada pela seguinte expressão:
$$
\\
$$
\begin{equation*}
    V(F) = 4,39 \times 10^{-3} \times F - 0,0057
\end{equation*}
$$
\\
$$
Manipulando a equação acima para que fique em função de <i>F</i> e explicitando que <i>V</i> é função do tempo, temos que:
$$
\\
$$
\begin{equation*}
    F = \frac{V(t) + 0,0057}{4,39 \times 10^{-3}}
\end{equation*}
$$
\\
$$
Considerando que o eixo possa ser considerado por um sistema massa-mola não-amortecido com vibrações livres, e sendo <i>m</i> e <i>k</i> a massa e a rigidez do eixo, respectivamente, tem-se que:
$$
\\
$$
\begin{equation*}
    m\cdot\ddot{u} + k\cdot u = 0  \;\;\;\;\;\; (1)
\end{equation*}
$$
\\
$$
Se o eixo tem seção reta circular de 7853,98 <i>mm²</i>, com 1000 <i>mm</i> de comprimento e é fabricado com aço <i>SAE 1045</i>, cujas densidade e módulo de elasticidade são  7,87 <i>g/cm³</i> e 190 <i>GPa</i>, respectivamente, temos que:
$$
\\
$$
\begin{equation*}
    k = \frac{A \times E}{L}
\end{equation*}
    
\begin{equation*}
    k = \frac{\frac{7853,98}{10^{6}} \times 190 \times 10^{9}}{\frac{1000}{10^{3}}}
\end{equation*}
    
<center>    
    <i>k</i> = 1492,25 <i>N</i>/<i>m</i>     
</center>
$$\\
\\$$
\begin{equation*}
    m = A \times L \times \rho
\end{equation*}
<br>\begin{equation*}
    m = \frac{7853,98}{10^{6}} \times \frac{1000}{10^{3}} \times \frac{7,85 \times 10^{6}}{10^{3}}
\end{equation*}

<center>
    <i>m</i> = 61,65 <i>kg</i>
</center>

$$
\\
\\
$$
Agora, retornando à equação (1), podemos escrever que:

\begin{equation*}
u = - \frac{m\cdot\ddot{u}}{k}
\end{equation*}

\begin{equation*}m\cdot\ddot{u} = F \end{equation*}

\begin{equation*}u = - \frac{F}{k} \end{equation*}

\begin{equation*}
u(V) = - \frac{V(t) + 0.0057}{k \times 4,39 \times 10^{-3}}
\end{equation*}
$$
\\
\\
$$
Pode-se, agora, converter o sinal dado em tensão elétrica para todo o domínio do tempo em um sinal dado em deslocamento, como mostra o gráfico abaixo.
</br>
</p>

```python
import matplotlib.pyplot as plt

t = np.arange(0,20,0.001)

sinal_V = 0.006*np.sin(2*t) + 0.0015*np.cos(2*t) + 0.05*np.sin(6*t) + 0.02*np.sin(6*t) + 0.25*np.sin(8*t) + 0.15*np.cos(8*t)

sinal_U = []

for i in range(0, len(sinal_V)):
    
    valor_instantaneo_sinal_V = -(sinal_V[i] + 0.0057)/(1492.25*4.39)
    
    sinal_U.append(valor_instantaneo_sinal_V)

plt.plot(t, sinal_U)
plt.xlabel('Tempo (s)')
plt.ylabel('Milímetros (mm)')
plt.title('Sinal adquirido por uma célula de carga acoplada a um eixo sob rotação')
```

<h2><b>Capítulo 3: <i>Hardware</i> de Aquisição de Dados (DAQ)</b></h2>
<p style='text-align: justify;'>
<br>
    Um <i>hardware</i> de aquisição de dados é um dispositivo eletro-eletrônico que transforma sinais analógicos adquiridos por um transdutor em sinais digitais. Dentro da placa DAQ, complexos circuitos elétricos funcionam em conjunto de forma a realizar uma filtragem analógica do sinal adquirido, bem como barramentos elétricos para minimizar possíveis interferência eletromagnéticas externas; os circuitos internos e seu funcionamento, no entanto, foge do escopo dessa apostila. É dentro da placa DAQ que se encontra o conversor analógico-digital, conhecido como conversor A/D ou ADC. De forma simplificada, um conversor A/D transforma um sinal no domínio do tempo em um sinal no domínio da frequência e cada dado passa a ser representado por pulsos de uma onda quadrada com aplitude e período constantes. Cada um desses pulsos representa uma unidade básica de informação, o <i>bit</i>, e pode assumir somente dois valores: 0 ou 1. Dependendo da memória utilizada, serão necessários entre 4 e 64 bits para representar um número binário.
</br>
<br>
No <b>exemplo</b> abaixo, é simulada uma conversão dos dados adquiridos no exemplo anterior, sendo apresentados somente as 250 primeiras amostras a fim de não poluir a visualização da apostila. Nele, é utilizada uma memória de 8 <i>bits</i>.

<h3>Exemplo:</h3>
</br></p>


```python
import numpy as np
import matplotlib.pyplot as plt
import struct

def float_to_bin(num):
    return format(struct.unpack('!I', struct.pack('!f', num))[0], '064b')

binary = []

for elements in sinal_U:
    
    binary_number = float_to_bin(elements)
    binary.append(binary_number)
    
print(binary[0:249])
```

<p style='text-align: justify;'>
<br>
Uma vez que esses dados analógicos foram convertidos em dados digitais, é possível convertê-los para números decimais novamente a fim de serem mais fáceis de serem interpretados pela linguagem humana. Assim, para as 250 primeiras amostras adquiridas para o exemplo anterior, agora transformadas para decimais, temos que: 
</br>
</p>

```python
import struct

def bin_to_float(binary):
    return struct.unpack('!f',struct.pack('!I', int(binary, 2)))[0]

bin_dec_sinal_U = []

for elements in binary:
    
    decimal_sinal_U = bin_to_float(elements)
    bin_dec_sinal_U.append(decimal_sinal_U)
    
print(bin_dec_sinal_U[0:249])
```

<h3><b>Erros de medida na aquisição de dados</b></h3>
<br>
Durante a aquisição de dados, alguns erros são passíveis de ocorrer, podendo eles serem de origem humana ou em função dos instrumentos de medida. Erros de natureza humana são bastante comuns e envolvem descuidos e limitações dos nossos biosensores. Já erros decorrentes de aparelhos de medição envolvem limitações devido a sua fabricação, manutenção e cuidados recorrentes, e são mensurados através da <b>resolução</b> do instrumento. Quanto maior a resolução de um aparelho, tanto maior será a sua <b>confiabilidade</b> quanto aos dados adquiridos. E, da mesma forma que nos instrumentos de medidas, nos sistemas de aquisição de dados esses erros também podem ocorrer. Tais erros podem-se dar de três formas: a primeira, pela propagação de erros ocorridos durante a medição; o segundo, por interferências e oscilações nos circuitos do conversor A/D; e o terceiro devido a erros de natureza numérica, chamados de <b>erros numéricos</b>. Tais erros ocorrem devido a aproximações realizadas pela placa DAQ durante a aquisição de dados e são inerentes dos processos numéricos, como ilustra o <b>exemplo</b> abaixo. Nele, é calculado o <b>erro quadrático médio</b> decorrente da conversão do sinal analógico original e a sua versão digital que foi transformada de números binários para decimais.
</br>
<h3>Exemplo:</h3>
</p>

```python
def erro_quadratico_medio(a, b):
    
    eqm = 0
    
    for i in range(0, len(a)):
        
        eqm += (a[i]-b[i])**2
        
    eqm = eqm/len(a)
    
    print('O erro quadrático médio entre o sinal analógico e o sinal digital para os deslocamentos é de ', eqm)
    return eqm

eqm = erro_quadratico_medio(sinal_U, bin_dec_sinal_U)
```

<p style='text-align: justify;'>
<br>
Embora o erro acima seja demasiadamente pequeno, ele existe e pode ser aumentado facilmente com a <b>propagação de erros</b> decorrentes do manuseio, manutenção, cuidados e resoluções do transdutor e do conversor A/D. Assim, para medições mais confiáveis, requere-se não somente um aparato instrumental de qualidade como também boas práticas de instrumentação. 
</br>
</p>

<h2><b>Capítulo 4: <i>Software</i> de Aquisição de Dados (DAQ)</b></h2>
<br>
Um <i>software</i> de aquisição de dados é um programa para manipulação, gerenciamento e monitoramento dos dados que são adquiridos. Nele, é possível realizar a visualização dos dados em tempo real, além de manipular e condicionar esses dados ou, até mesmo, salvá-los para análises futuras. Nessa apostila, será apresentado brevemente algumas das características e funcionalidades do <i>software LabView</i>, já citado anteriormente.
</br>
</p>

<p style='text-align: justify;'>
<br>
O <i>LabView</i> é um software desenvolvido pela <i>National Instruments Inc.</i> (NI), tendo sua primeira versão lançada em 1986 para os computadores da linha <i>Macintosh</i>. De lá pra cá, o programa foi atualizado e disponibilizado para diferentes sistemas operacionais, como <i>Windows</i> e <i>Linux</i>, estando agora na sua vigésima versão, o <i>LabView 2020</i>. Há, inclusive, uma versão gratuita para aplicações não-comerciais e que oferece todas as funcionalidades da versão profissional. A linguagem de programação utilizada pelo <i>LabView</i> é a linguagem G, que remete a uma programação <b>gráfica</b>, programada mediante a utilização de diagramas de blocos e sendo, portanto, uma linguagem de alto nível.
